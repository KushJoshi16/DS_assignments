```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
%matplotlib inline
```

- ## Get all the Earth meteorites that fell before the year 2000

```python
df = pd.read_csv('MeteoriteDataset.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['year'] = pd.to_datetime(df['year'],errors = 'coerce')
```

```python
df['year'].fillna(pd.Timestamp.min,inplace=True)
```

```python
df[df['year'].isnull()]
```

```python
plt.figure(figsize = (10,5))
plt.hist(df['year'].dt.year,bins=50)
plt.axvline(x=df['year'].dt.year.min(),color = 'g')
plt.axvline(x = 2000, color = 'r', linestyle = 'dotted')
plt.title('Plot for spawn rate (min and 2000 marked)')
plt.show()
```

```python
df[df['year'].dt.year<2000]
```

- ## Get all the earth meteorites co-ordinates who fell before the year 1970

```python
RequiredCoordinates = df[df['year'].dt.year<1970][['year','reclong','reclat','Point Coordinates']]
```

```python
plt.figure(figsize = (10,5))
plt.hist(df['year'].dt.year,bins=50)
plt.axvline(x=df['year'].dt.year.min(),color = 'g')
plt.axvline(x = 1970, color = 'r', linestyle = 'dotted')
plt.title('Plot for Year distributin (min and 1970 marked)')
plt.show()
```

```python
plt.figure(figsize = (11,5.5))
plt.scatter(x = RequiredCoordinates['reclong'],y= RequiredCoordinates['reclat'],c = RequiredCoordinates['year'])
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earth meteorites coordinates')
plt.show()
```

```python
RequiredCoordinates.drop(['year'],axis=1,inplace=True)
```
