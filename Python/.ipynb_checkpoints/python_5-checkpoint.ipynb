```python
import json
from urllib.request import urlopen
import pandas as pd
```

```python
url = 'http://api.tvmaze.com/singlesearch/shows?q=westworld&embed=episodes'
```

```python
# df = pd.read_json(url)
```

```python
response = urlopen(url)
```

```python
data = json.loads(response.read())
```

```python
data
```

```python
pd.json_normalize(data)
```

```python
data['_embedded']
```

```python
pd.json_normalize(data['_embedded'])
```

```python
df = pd.json_normalize(data['_embedded'],record_path="episodes")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df['summary'].unique()
```

```python
df['summary'].replace(to_replace=r"<(/?.+?/?)>", value="", regex=True,inplace=True)
```

```python
df['summary'].unique()
```

```python
df.info()
```

```python
df.columns
```

```python
df.rename(columns={'rating.average':'average rating','image.medium':'medium image link','image.original':'Original image link'},inplace=True)
```

```python
df.head(2)
```

```python
df.drop(['_links.self.href', '_links.show.href'],axis=1,inplace=True)
```

```python
df.info()
```

```python
df.head()
```

```python
df['airdate'].unique()
```

```python
df['airdate'] = pd.to_datetime(df['airdate'])
```

```python
df['airtime'].unique()
```

```python
df['airtime'] = pd.to_datetime(df['airtime'], format="%H:%M").apply(lambda x: x.strftime("%I:%M:%S %p"))
```

```python
df
```

```python
df.to_csv("Episodes.csv",index=False)
```
