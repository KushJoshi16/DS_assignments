```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
url = 'https://data.wa.gov/api/views/f6w7-q2d2/rows.csv?accessType=DOWNLOAD'
df = pd.read_csv(url)
```

```python
df.info()
```

- ## Get all the cars and their types that do not qualify for clean alternative fuel vehicle

```python
df.columns
```

```python
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].unique()
```

```python
CAFV_DF = df[df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'] != 'Clean Alternative Fuel Vehicle Eligible'][['Make', 'Model','Electric Vehicle Type']]
```

```python
CAFV_DF
```

- ## Get all TESLA cars with the model year, and model type made in Bothell City.

```python
BothellCityCars_df = df[(df['Make'] == 'TESLA') & (df['City'] == 'Bothell')]
```

```python
BothellCityCars_df
```

- ## Get all the cars that have an electric range of more than 100, and were made after 2015

```python
RangeDate_df = df[(df['Electric Range'] > 100) & (df['Model Year'] > 2015)]
```

```python
RangeDate_df
```
