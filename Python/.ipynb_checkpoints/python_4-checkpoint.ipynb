
- Name of Earth Meteorite - string 
- id - ID of Earth Meteorite - int 
- nametype - string 
- recclass - string
- mass - Mass of Earth Meteorite - float 
- year - Year at which Earth Meteorite was hit - datetime format 
- reclat - float 
- recclong - float
- point coordinates - list of int


```python
import json
from urllib.request import urlopen
import pandas as pd
```

```python
url = 'https://data.nasa.gov/resource/y77d-th95.json'
```

```python
df = pd.read_json(url)
```

```python
df['Point Coordinates'] = df.geolocation.apply(pd.Series)['coordinates']
```
