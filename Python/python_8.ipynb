```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings("ignore")
%matplotlib inline
```

- ## Get all the overall ratings for each season and using plots compare the ratings for all the seasons, like season 1 ratings, season 2, and so on.

```python
df = pd.read_csv("Episodes.csv")
```

```python
df
```

```python
df['season'].unique()
```

```python
df['average rating'].unique()
```

```python
df.groupby('season').mean('average rating')
```

```python
plt.figure(figsize = (7,5))
sns.barplot(y = 'average rating', x = 'season',data=df.groupby('season',as_index=False).mean('average rating'))
plt.yticks(size=8)
plt.show()
```

```python
df.groupby('season',as_index=False).mean('average rating')[['season','average rating']]
```

- ## Get the episode name from each season with the highest and lowest rating

```python
df.groupby('season').agg(minimum_rating = ('average rating', 'min'), maximum_rating = ('average rating', 'max'))
```

```python
df.loc[df.groupby('season')['average rating'].idxmin()][['season','name','average rating']]
```

```python
df.loc[df.groupby('season')['average rating'].idxmax()][['season','name','average rating']]
```

- ## Get the summary for the most popular ( ratings ) episode in every season

```python
df.loc[df.groupby('season')['average rating'].idxmax()][['season','summary']]
```
